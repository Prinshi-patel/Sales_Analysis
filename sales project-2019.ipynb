```python
data = pd.read_csv("./all_data.csv")

print(data)
```

```python
data.isnull().sum()
```

task : drop null values

```python
all_data = data.dropna(how = "all")
```

```python
print(all_data)
```

Task : Adding Month column

```python
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data.head()
```

```python
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
print(all_data)
```

```python

temp_data = all_data[all_data["Order Date"].str[0:2] == 'Or' ]
print(temp_data.head())
```

```python

all_data["Month"] = all_data["Month"].astype("int32")
print(all_data.head())
```

```python
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
```

TAsk :Adding    sales column

```python
all_data["sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
print(all_data.head())
```

TAsk : Adding  city column

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(",")[2].split(" ")[1]
all_data["city"]= all_data["Purchase Address"].apply(lambda x: f"{get_city(x)} ({get_state(x)}) ")
```

```python
results = all_data.groupby("Month").sum(numeric_only= True)
print(results)
```

q_1 : what was the best month for sales? How much was earned that month?

```python
import matplotlib.pyplot as plt 
months = range(1,13)
plt.bar(months,results["sales"],color= "g")
plt.xticks(months)
plt.xlabel("month name")
plt.ylabel("Sales in usd ($)")
plt.legend()
plt.show()
```

```python
df = all_data[all_data["Order ID"].duplicated(keep=False)]
df.head(20)
```

Task : Adding column of product group & drop Duplicates

```python
df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x : ",".join(x))
df.head()
```

```python
df = df[["Order ID","Grouped"]].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2)))

print(count)
for key, value in count.most_common(10):
    print(key,value)
```

q -5 what product sold the most? why do you think it sold most?

```python
pd_group = all_data.groupby("Product")

```

```python
quantity_order = pd_group["Quantity Ordered"].sum()
products = [product for product , df in pd_group]
plt.bar(products,quantity_order)
plt.xlabel("quantity Order")
plt.ylabel("Product")
plt.xticks(products, rotation = "vertical",size = 10)
plt.show()
```

```python
prices = all_data.groupby("Product")["Price Each"].mean()
print(prices)
```

```python
fig,ax1 = plt.subplots()
ax2 =ax1.twinx()
ax1.bar(products,quantity_order,color ="g")
ax2.plot(products, prices, '-b')

ax1.set_xlabel("Product Name")
ax1.set_ylabel("Quantity Order", color = "g")
ax2.set_ylabel("Price ($)",color = "b")
ax1.set_xticklabels(products, rotation= "vertical", size=6)
plt.show()
```
